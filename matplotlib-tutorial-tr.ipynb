# Matplotlib Nedir? <a id="matpltlib-nedir"></a>
Matplotlib, Python'da statik, animasyonlu ve etkileşimli veri görselleştirmesi yapmak için kullandığımız kütüphanedir.

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Figür Nedir? <a id="figur-nedir"></a>

<img src="https://matplotlib.org/_images/anatomy.png">

### Figür (*Figure*):
Figür grafiğin sınırlarını(*axes*) ve çerçeveyi(*canvas*) kapsar.

```python
fig = plt.figure()  # boş figür. çıktısı yok.
fig, ax = plt.subplots()  # içinde tek grafik olan figür
fig, axs = plt.subplots(2, 2)  # 4 grafik içeren figür.
```

### Grafik (*Axes*):
* Bir figür birden fazla grafik içerebilir fakat her grafik tek bir figürde bulunabilir.
* Grafik 2 (3 boyutlu çizimler için 3) eksen objesi içerir.
* Her grafiğin başlığı ve eksen etiketleri vardır.

## Basit Çizgi Grafiği Çizimi <a id="cizgi-grafigi"></a>

```python
fig , ax = plt.subplots() #figür oluşturma.
ax.plot([1,2,3,4] , [1,4,2,3])
```

```python
plt.plot([1,2,3,4] , [1,4,2,3]) # tek satırda grafik çizimi.
```

### Örnek Grafikler
**object-oriented grafik çizimi (interaktif olmayan projeler için daha çok tercih edilir):**

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()  # Figür ve grafik oluşturma
ax.plot(x, x, label='linear')  # Veriyi grafik üzerine çizme
ax.plot(x, x**2, label='quadratic') 
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')  # X eksenini etiket eklemek
ax.set_ylabel('y label')  # Y eksenine etiket eklemek
ax.set_title("Simple Plot")  # Grafiğe başlık eklemek
ax.legend()  # Açıklama yazısı
```

**pyplot grafik çizimi:**

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

# Pyplot ile Grafik Çizimi <a id="pyplot"></a>

* `plt.plot(x,y)` : x ve y'nin grafiğini çizer.
* `plt.show()` : grafiği gösterir.
* `plt.title(title)` : grafiğe başlık ekler.
* `plt.xlabel(xlabel)` : x eksenine etiket ekler
* `plt.ylabel(ylabel)` : y eksenine etiket ekler

```python
plt.plot([1,2,3,4] , [1,4,9,16])
plt.title("örnek grafik")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.show()
```

```python
plt.plot([1,2,3,4] , [1,4,9,16] ,  color='green', marker='o', linestyle='dashed',
     linewidth=2, markersize=8)
plt.axis([0,6,0,20]) # (xmin,xmax,ymin,ymax)
plt.show()
```

# Kategorilenmiş Değerlerin Grafiği <a id="kategorili-grafik"></a>

* `plt.bar(x,y)` : sütun grafiği
* `plt.scatter(x,y)` : nokta grafiği
* `plt.plot(x,y)` : çizgi grafiği
* `plt.hist(x,y)` : histogram çizimi

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))  #figsize figürün boyutunu belirler.

plt.subplot(131)    # 1 satır 3 sütun 1. grafik
plt.bar(names, values)
plt.subplot(132)    # 1 satır 3 sütun 2. grafik
plt.scatter(names, values)
plt.subplot(133)    # 1 satır 3 sütun 3. grafik
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

# HeatMap Örneği <a id="heatmap"></a>

```python
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest)

# We want to show all ticks...
ax.set_xticks(np.arange(len(farmers)))
ax.set_yticks(np.arange(len(vegetables)))
# ... and label them with the respective list entries
ax.set_xticklabels(farmers)
ax.set_yticklabels(vegetables)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```
